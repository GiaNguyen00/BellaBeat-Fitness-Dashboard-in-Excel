```python
import pandas as pd
```

```python
df1 = pd.read_csv('Data/Steps Minute Record.csv')
df1
```

```python
temp = df1.groupby('Day of Week')['Steps'].sum()
output_df1 = pd.DataFrame(temp,index=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
output_df1
```

```python
df2 = pd.read_csv('Data/Sleep Records.csv')
df2 = df2[['DayofWeek','TotalHrInBed','TotalHrAsleep']]
temp = df2.groupby('DayofWeek')[['TotalHrInBed','TotalHrAsleep']].sum()
output_df2 = pd.DataFrame(temp,index=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
output_df2
```

```python
df3 = pd.read_csv('Data/Intensity Records.csv')
df3 = df3[['Hour','AverageIntensity']]
temp = df3.groupby('Hour')['AverageIntensity'].mean()
output_df3 = pd.DataFrame(temp,index=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
output_df3
```

```python
df4 = pd.read_csv('Data/Daily Activity Records.csv')
df4 = df4[['Hour','TotalSteps','Calories','VeryActiveDistance','ModeratelyActiveDistance','LightActiveDistance','SedentaryActiveDistance']]
temp = df4.groupby('Hour')[['TotalSteps','Calories','VeryActiveDistance','ModeratelyActiveDistance','LightActiveDistance','SedentaryActiveDistance']].sum()
output_df4 = pd.DataFrame(temp,index=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
output_df4
```

```python
# Assuming you have two dataframes df1 and df2
output_df1.join(output_df2).join(output_df3).join(output_df4).to_csv('Data/Summary.csv')
```
